### Q1: First name only
https://www.interviewquery.com/questions/first-names-only

```python
import pandas as pd

users = {"user_id" : [1034, 9430, 7281, 5264, 8995], 
         "full_name" : ["James Emerson", "Fiona Woodward", "Alvin Gross", 
                        "Deborah Handler", "Leah Xue"]}

users_df = pd.DataFrame(users)
users_df
```

```python
users_df['first_name'] = users_df["full_name"].apply(lambda x : x.split(" ")[0])
users_df.drop(["full_name"], axis = 1, inplace = True)
```

```python
users_df
```

### Q2: Complete Address
https://www.interviewquery.com/questions/complete-addresses

```python
import pandas as pd
addresses = {"address": ["4860 Sunset Boulevard, San Francisco, 94105", "3055 Paradise Lane, Salt Lake City, 84103", "682 Main Street, Detroit, 48204", "9001 Cascade Road, Kansas City, 64102", "5853 Leon Street, Tampa, 33605"]}
cities = {"city": ["Salt Lake City", "Kansas City", "Detroit", "Tampa", "San Francisco"], "state": ["Utah", "Missouri", "Michigan", "Florida", "California"]}
df_addresses = pd.DataFrame(addresses)
df_cities = pd.DataFrame(cities)
```

```python
df_addresses
```

```python
df_cities
```

```python
df_addresses["street"] = df_addresses["address"].apply(lambda x : x.split(",")[0])
df_addresses["city"] = df_addresses["address"].apply(lambda x : x.split(",")[1].strip())
df_addresses["zipcode"] = df_addresses["address"].apply(lambda x : x.split(",")[2])
df_addresses
```

```python
df = df_addresses.copy()
data = df.merge(df_cities, how = "left", on = "city")
data
```

```python
data["address"] = data["street"] + ', ' + data["city"] + ', ' + data["state"] + ', ' + data["zipcode"]
data.drop(["street", "city", "zipcode", "state"], axis = 1)
```

##### 官方解答

```python
df_addresses = pd.DataFrame(addresses)
df_addresses[["street", "city", "zipcode"]] = df_addresses["address"].str.split(", ", expand = True)
df_addresses
```

```python
df_addresses = df_addresses.merge(df_cities, on = "city")
df_addresses["address"] = df_addresses[["street", "city", "state", "zipcode"]].agg(', '.join, axis = 1)
df_addresses.drop(["street", "city", "state", "zipcode"], axis = 1)
```

### Q3: fillna with median
https://www.interviewquery.com/questions/impute-median


```python
from numpy import nan
import numpy as np

cheeses = {"Name": ["Bohemian Goat", "Central Coast Bleu", "Cowgirl Mozzarella", "Cypress Grove Cheddar", 
                    "Oakdale Colby"], "Price" : [15.00, np.nan, 30.00, np.nan, 45.00]}

cheese_df = pd.DataFrame(cheeses)
cheese_df
```

```python
cheese_df.fillna(cheese_df["Price"].median())
```

### Q4: Good Grades and Favorite Colors
https://www.interviewquery.com/questions/good-grades-and-favorite-colors

```python
students = {"name" : ["Tim Voss", "Nicole Johnson", "Elsa Williams", "John James", "Catherine Jones"], "age" : [19, 20, 21, 20, 23], "favorite_color" : ["red", "yellow", "green", "blue", "green"], "grade" : [91, 95, 82, 75, 93]}
students_df = pd.DataFrame(students)
students_df
```

```python
# Write a function to select only the rows where the student's favorite color is green or red and their grade is above 90.
students_df[(students_df["favorite_color"].isin(["green", "red"])) & (students_df["grade"] >= 90)]
```

```python
# or
students_df[students_df["favorite_color"].str.contains("red|green")]
```

### Q4: Over 100 Dollars
https://www.interviewquery.com/questions/over-100-dollars

```python
transactions = {"transaction_id" : [1, 2, 3, 4, 5], "product_id" : [101, 102, 103, 104, 105], "amount" : [3, 5, 8, 3, 2]}
products = {"product_id" : [101, 102, 103, 104, 105], "price" : [20.00, 21.00, 15.00, 16.00, 52.00]}
df_transactions = pd.DataFrame(transactions)
df_products = pd.DataFrame(products)
df_transactions
```

```python
df_products
```

```python
# return a dataframe containing every transaction with a total value of over $100. 
#Include the total value of the transaction as a new column in the dataframe.
df = df_transactions.merge(df_products, how = 'left', on = 'product_id')
df["total"] = df["amount"] * df["price"]
df[df["total"] >= 100]
```
