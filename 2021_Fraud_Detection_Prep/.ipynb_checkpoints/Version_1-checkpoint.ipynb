```python
import pandas as pd
```

```python
data = pd.read_csv("creditcard.csv", sep = ",")
data.head(3)
```

```python
data.info()
```

```python
data['Class'] = data['Class'].astype("str")
```

```python
dist = data.groupby(["Class"]).agg({"Time":"count", "Amount":"mean"}).reset_index()
dist.columns = ["Class", "Number of case", "Average amount"]
dist
#funnel_new.groupby(['merchant_id']).size().reset_index(name = 'Number of checkout data')
```
