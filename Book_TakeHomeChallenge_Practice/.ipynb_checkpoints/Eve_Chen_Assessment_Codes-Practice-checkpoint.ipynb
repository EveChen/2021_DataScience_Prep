### Question 2: Customer repeat purchase behavior
1. Define "repeat behavior": if a customer bought more than once in six months, we regard this customer as the repeated customer.
2. Made a new user purchase behavior table, labeled whether the user has mutiple buying behaviors
3. Check the correlation between each variables, drop those with high correlation before we apply the random forest model.
4. Apply the random forest model to the new dataset, identify the feature importance.
5. Dive deep to the most important feature to extract more insights

```python
# Given it's about customer "purchase" bahavior, we used the "loan" table instead of the funnel table
loan_new.head(3)
```

```python
# Create a new fico table for futher data join
fico = loan_new[['user_id', 'fico']]
fico = fico.drop_duplicates()
```

```python
# Create the new user purchase behavior table by adding new features
user = loan_new.groupby(['user_id']).agg({'checkout_id':'count', 'merchant_id': 'count', 'loan_amount':'mean'}).reset_index()
user.columns = ['user_id', '#checkouts', '#merchants', 'avg loan amount']

# Add Fico as one variable
user = user.merge(fico, on = 'user_id', how = 'left')
user.head(3)
```

```python
# Add marketing channel as one of the variables
user = user.merge(mkt, on = 'user_id', how = 'left')

# Labeled 'repeated' = 1 if the customer purchase more than once
user.loc[user['#checkouts'] > 1, 'repeated'] = 1
user.loc[user['#checkouts'] == 1, 'repeated'] = 0
user['repeated'] = user['repeated'].astype('object')
user.head(3)
```

```python
# Create an user approve table to check the approval rate
user_approved = funnel_new.groupby(['user_id']).agg({'checkout_id':'count', 'approved':'sum'}).reset_index()
user_approved['approved_rate'] = round(user_approved['approved'] / user_approved['checkout_id'], 2)
user = user.merge(user_approved, on = 'user_id')
user = user.drop(['date', 'checkout_id', 'approved'], axis = 1)
user.head(3)
```

```python
# Add marketing channel as one of the variables
user.loc[user['last_touch_marketing_channel'].isnull(), 'mkt_channel'] = 0
user.loc[user['last_touch_marketing_channel'] == 'Direct Load', 'mkt_channel'] = 1
user.loc[user['last_touch_marketing_channel'] == 'Display', 'mkt_channel'] = 2
user.loc[user['last_touch_marketing_channel'] == 'Email', 'mkt_channel'] = 3
user.loc[user['last_touch_marketing_channel'] == 'Organic Search', 'mkt_channel'] = 4
user.loc[user['last_touch_marketing_channel'] == 'Paid Social', 'mkt_channel'] = 5
user.loc[user['last_touch_marketing_channel'] == 'Push', 'mkt_channel'] = 6

user['last_touch_marketing_channel'] = user['last_touch_marketing_channel'].fillna('unknown')
user['mkt_channel'] = user['mkt_channel'].astype('object')

user.head(3)
```

#### Use heatmap to check the correlation between each variables
--> Remove the highly correlated variables for not heavily influencing model performance

```python
# We found the #checkouts and #merchants did have a high correlation. Drop those two variables
import seaborn as sns
user_df = user.drop(['user_id', 'repeated'], axis = 1)
Var_Corr = user_df.corr()
# plot the heatmap and annotation on it
sns.heatmap(Var_Corr, xticklabels=Var_Corr.columns, yticklabels=Var_Corr.columns, annot=True)
```

#### Apply Random Forest model to the new user purchase behavior table

```python
# Load in needed python packages
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Identified the x and y variables
y = np.array(user['repeated'])
y = y.astype('int')
x_var = user[['avg loan amount','fico', 'approved_rate', 'mkt_channel']] #remove #merchants
x = np.array(x_var)
column_list = list(user.columns)

# Apply the random forest model
rnd_clf = RandomForestClassifier(n_estimators=500, n_jobs=-1, random_state=42)
rnd_clf.fit(x, y)
```

```python
# Calculate feature importances
importances = rnd_clf.feature_importances_

# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Rearrange feature names so they match the sorted feature importances
names = [x_var.columns[i] for i in indices]

# Barplot: Add bars
plt.bar(range(x.shape[1]), importances[indices])
# Add feature names as x-axis labels
plt.xticks(range(x.shape[1]), names, rotation=20, fontsize = 8)
# Create plot title
plt.title("Feature Importance")
# Show plot
plt.show()
```

```python
def calculate_df(df, dim):
    total_checkout = len(df)
    appli = df[df['application'] == 1]
    appro = df[df['approved'] == 1]
    conf = df[df['confirmed'] == 1]
    if dim == 'fico_tranche':
        mer_funnel = df.groupby([dim]).agg({'loan_amount':'mean', 'loan_return':'mean'}).reset_index()
    else:
        mer_total = df.groupby([dim])['checkout_id'].size().reset_index(name = 'Total number of checkout')
        mer_appli = appli.groupby([dim])['application'].size().reset_index(name = 'Number of application')
        mer_appro = appro.groupby([dim])['approved'].size().reset_index(name = 'Number of approved')
        mer_conf = conf.groupby([dim])['confirmed'].size().reset_index(name = 'Number of confirmed')
        mer_funnel = mer_total.merge(mer_appli, on = dim)
        mer_funnel = mer_funnel.merge(mer_appro, on = dim)
        mer_funnel = mer_funnel.merge(mer_conf, on = dim)
        
    return mer_funnel
```
