```python
def extract(obj, keys, nullable = True):
    for k in keys:
        if isinstance(k, str):
            obj = obj.get(k)
            if obj is None:
                if nullable: #Q: nullable?
                    return None
                else:
                    raise KeyError()
        elif isinstance(k, int):
            if len(obj) <= k:
                if nullable:
                    return None
                else:
                    raise IndexError()
            else:
                obj = obj[k]
        else:
            raise Exception('Unknown key: {key} with type {type_'.format(key=repr(k), type_=type(k)))
    return obj
```

### Q6: Write a function to combine and sanitize "StreetNumber", "StreetName", "StreetSuffix", "StreetDirPrefix" into a full address. Ensure that you validate each field

```python
# 100筆資料中只有五筆資料有 full address
def sanitize_address(data):
    full_address = []
    for i in range(len(data)):
        if 'StreetNumber' in data[i] and 'StreetName' in data[i] and 'StreetSuffix' in data[i] and 'StreetDirPrefix' in data[i]:
            full_address.append(data[i]['StreetNumber'] + ' ' + data[i]['StreetName'] + ' ' +
                               data[i]['StreetSuffix'] + ' ' + data[i]['StreetDirPrefix'])
    return full_address
```

```python
sanitize_address(data)
```

```python
# Q: 如果問題換成「若沒有該 field 的資訊就加上 "?"，但是要 print 出所有 100 筆地址」，怎麼做!?
# 下面是員億寫的醜醜 codes...又難看又有 bug 嗚嗚
def sanitize_address(data):
    full_address = []
    for i in range(len(data)):
        if isinstance(data[i]['StreetNumber'], int):
            street_num = data[i]['StreetNumber']
        else:
            street_num = '?'
        
        if isinstance(data[i]['StreetName'], str):
            street_name = data[i]['StreetName']
        else:
            street_name = '?'
            
        if isinstance(data[i]['StreetSuffix'], str):
            street_suffix = data[i]['StreetSuffix']
        else:
            street_suffix = '?'
            
        if isinstance(data[i]['StreetDirPrefix'], str):
            street_dir = data[i]['StreetDirPrefix']
        else:
            street_dir = '?'
        
            
        full_address.append(street_num + ' ' + street_name + ' ' + street_suffix + ' ' + street_dir)
    return full_address
```

### Q7: Write a function to validate and sanitize agent names (ListAgentFullName and CoListAgentFullName)

```python
def sanitize_name(data):
    agent_name = []
    for i in range(len(data)):
        if 'ListAgentFullName' in data[i] and 'CoListAgentFullName' in data[i]:
            agent_name.append([data[i]['ListAgentFullName'], data[i]['CoListAgentFullName']])
    return pd.DataFrame(agent_name, columns = ['ListAgentFullName', 'CoListAgentFullName'])
```

```python
sanitize_name(data)
```
