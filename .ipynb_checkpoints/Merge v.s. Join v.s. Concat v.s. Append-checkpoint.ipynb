```python
import pandas as pd
import numpy as np
```

### pd.concat(list)
Only on **list** or **dict**. very useful for concatenating different files like
* frames = [ process_your_file(f) for f in files ]
* result = pd.concat(frames)

### df1.append(df2)
Only on **Series** or **DataFrame**. These methods actually predated concat. They concatenate along axis=0, namely the index

```python
df1 = pd.DataFrame(
         {
            "A": ["A0", "A1", "A2", "A3"],
            "B": ["B0", "B1", "B2", "B3"],
            "C": ["C0", "C1", "C2", "C3"],
            "D": ["D0", "D1", "D2", "D3"],
        },
        index=[0, 1, 2, 3],
    )
    

df2 = pd.DataFrame(
        {
            "A": ["A4", "A5", "A6", "A7"],
            "B": ["B4", "B5", "B6", "B7"],
            "C": ["C4", "C5", "C6", "C7"],
            "D": ["D4", "D5", "D6", "D7"],
        },
        index=[4, 5, 6, 7],
    )
    

df3 = pd.DataFrame(
        {
            "A": ["A8", "A9", "A10", "A11"],
            "B": ["B8", "B9", "B10", "B11"],
            "C": ["C8", "C9", "C10", "C11"],
            "D": ["D8", "D9", "D10", "D11"],
        },
        index=[8, 9, 10, 11],
    )

df4 = pd.DataFrame(
        {
            "B": ["B2", "B3", "B6", "B7"],
            "D": ["D2", "D3", "D6", "D7"],
            "F": ["F2", "F3", "F6", "F7"],
        },
        index=[2, 3, 6, 7],
    )
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# Concat df1, df2, df3
pd.concat([df1, df2, df3])

# same as 
# df1.append([df2, df3])
```

```python
### inner join & outer join (default = outer join)
df1
```

```python
df4
```

```python
pd.concat([df1, df4]) # default = outer join & axis = 0

# same as
# df1.append(df4)
```

```python
pd.concat([df1, df4], axis = 1)
```

```python
pd.concat([df1, df4], join = "inner") # axis = 0
```

```python
pd.concat([df1, df4], join = "inner", axis = 1)
```

```python
# Reindex
df4.reindex(df1.index)
```

```python
pd.concat([df1, df4], ignore_index= True)

#same as
# df1.append(df4, ignore_index = True)
```
